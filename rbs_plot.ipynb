```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rbs_metacell = np.load('/Users/ckw/warehouse/metacell/stark/rbs_seacell.npy', allow_pickle=True).item()
rbs_metacell = pd.DataFrame(rbs_metacell).T
rbs_metacell.head(5)
rbs_metacell.columns
```

```python
# ['mean_purity', 'accuracy', 'global_score', 'WCOS', 'HWIS']
fig = plt.figure(figsize=(10, 10))
plt.plot(rbs_metacell.index, rbs_metacell['mean_purity'], 'o-', label='mean_purity')
plt.plot(rbs_metacell.index, rbs_metacell['accuracy'], 'o-', label='accuracy')
plt.plot(rbs_metacell.index, rbs_metacell['global_score'], 'o-', label='global_score')
plt.plot(rbs_metacell.index, rbs_metacell['WCOS'], 'o-', label='WCOS')
plt.plot(rbs_metacell.index, rbs_metacell['HWIS'], 'o-', label='HWIS')
plt.legend()
plt.grid()
plt.xlabel('metacell number')
plt.ylabel('score')
plt.title('RBS metacell evaluation')
plt.show()
```
